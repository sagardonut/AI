```python
A = np.array([[2, -3,  1],
              [1, -1,  2],
              [3,  1, -1]])

B = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A)   # find A inverse

X = A_inv @ B              # matrix multiplication

print("A inverse:\n", A_inv)
print("\nSolution vector X = [x, y, z]:\n", X)
```
